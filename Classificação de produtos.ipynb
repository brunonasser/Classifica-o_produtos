### Preenchendo os dados que faltam 

```python
df1['COMPOSICAO'].fillna(value='plastico',inplace=True)
```

```python
df1['FORMATO'].fillna(value='chaves',inplace=True)
```

```python
df1['PARA_LAVA_LOUCAS'].fillna(value='sei la',inplace=True)
```

```python
df1['PARA_MICRO_ONDAS'].fillna(value='numsei',inplace=True)
```

```python
df1['CAPACIDADE_(L)'] = df1.groupby(['COMPOSICAO','FORMATO'])['CAPACIDADE_(L)']\
    .transform(lambda x: x.fillna(x.mean()))
```

```python
df1['PESO'] = df1.groupby(['COMPOSICAO','FORMATO','TEM_FERRO_FUNDIDO'])['PESO']\
    .transform(lambda x: x.fillna(x.mean()))
```

```python
df1['TEMPO_GARANTIA'] = df1.groupby(['MARCA'])['TEMPO_GARANTIA']\
    .transform(lambda x: x.fillna(x.mean()))
```

```python
df1.dropna(axis=0,inplace=True)
```

## Convertendo variáveis categóricas em numéricas

```python
from sklearn.preprocessing import LabelEncoder
cols = ('COMPOSICAO','COR','FORMATO','MARCA','PARA_LAVA_LOUCAS','PARA_MICRO_ONDAS','TEM_FERRO_FUNDIDO','TEM_GRELHA','TEM_TAMPA','TIPO_PRODUTO','TIPO_WOK','ITEM_PRICE')
#aplicando LabelEncoder para as variáveis categóricas
for c in cols:
    lbl = LabelEncoder()
    lbl.fit(list(df1[c].values))
    df1[c] = lbl.transform(list(df1[c].values))
```

```python
df1.head(10)
```

```python
#dando uma tapinha no visual
df1.drop(['ITEM_ID'],axis=1,inplace=True)
```

## Escolha do modelo:

Como os dados apresentam a coluna INTERESTED completa é correto utilizar um algoritmo de aprendizagem supervisionada. O objetivo é classificar as panelas entre interessante ou não analisando diversas variáveis categóricas e numéricas. Um excelente modelo que se encaixa nessa situação é o modelo de Classificação por Florestas aleatórias. Este algoritmo tem a capacidade de entender qual a importancia de cada característica para a predição e prove o melhor caminho para isso.

### Treinando o modelo de Random Forest 

Precisamos primeiro dividir nossos dados em uma matriz X que contém os recursos para treinar, e uma matriz y com a variável alvo, neste caso, a coluna INTERESTED.

```python
X = df1[['ALTURA','CAPACIDADE_(L)','COMPOSICAO','COR','FORMATO','LARGURA','MARCA','PARA_LAVA_LOUCAS','PARA_MICRO_ONDAS','PESO','PROFUNDIDADE','TEMPO_GARANTIA','TEM_FERRO_FUNDIDO','TEM_GRELHA','TEM_TAMPA','TIPO_PRODUTO','TIPO_WOK','SESSION_ID','ITEM_PRICE']]
y = df1['INTERESTED']
```

```python
#dividindo entre dados de treino e teste:
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

#### Criando e treinando o modelo

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc=RandomForestClassifier(n_estimators=19)
```

```python
rfc.fit(X_train,y_train)
```

```python
print("Accuracy on training set: {:.3f}".format(rfc.score(X_train, y_train)))
print("Accuracy on test set: {:.3f}".format(rfc.score(X_test, y_test)))
```

Pelo score acima pode-se observar que o número de acertos para os dados de teste e treino foi próximo de 90%, número considerado como muito bom.

```python
rfc_pred=rfc.predict(X)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y,rfc_pred))
print('\n')
print(confusion_matrix(y,rfc_pred))
```
